```python
dice = [1,2,3,4,5,6]
prob = [1/21, 2/21, 3/21, 4/21, 5/21, 6/21]
```

```python
num_trial = 100
sample = np.random.choice(dice, num_trial, p=prob)
sample
```

```python
num_trial = 100
sample = np.random.choice(dice, num_trial, p=prob)

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
ax.hist(sample, density=True, align='left')
ax.set_xlabel('score')
ax.set_ylabel('relative frequency')
plt.show()
```

```python
num_trial = 10000
sample = np.random.choice(dice, num_trial, p=prob)

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
ax.hist(sample, density=True, align='left')
ax.set_xlabel('score')
ax.set_ylabel('relative frequency')
plt.show()
```
